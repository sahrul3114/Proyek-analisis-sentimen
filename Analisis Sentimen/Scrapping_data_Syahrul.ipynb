```python
# Mengimpor pustaka google_play_scraper untuk mengakses ulasan dan informasi aplikasi dari Google Play Store.
from google_play_scraper import app, reviews, Sort, reviews_all
```

```python
# Mengambil semua ulasan dari aplikasi Wondr by BNI dengan ID 'id.bni.wondr' di Google Play Store.
scrapreview = reviews_all(
    'id.bni.wondr',          # ID aplikasi
    lang='id',             # Bahasa ulasan (default: 'en')
    country='id',          # Negara (default: 'us')
    sort=Sort.MOST_RELEVANT, # Urutan ulasan (default: Sort.MOST_RELEVANT)
    count=3000             # Jumlah maksimum ulasan yang ingin diambil
)

print(f'Kamu berhasil mendapatkan Data ulasan aplikasi Wondr by BNI sebanyak {len(scrapreview)}')
```

```python
import pandas as pd
df = pd.DataFrame(scrapreview)
df.info()
```

```python
df_sample = df.sample(n=5000, random_state=42)
df_sample.to_csv('ulasan_aplikasi_wondr_5k.csv', index=False, encoding='utf-8')

print(f"yeayy, sudah berhasil menyimpan data menjadi 'ulasan_aplikasi_wondr_5k.csv' tinggal download deh")
```
